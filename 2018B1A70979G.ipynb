```
#OPTIONAL ASSIGNMENT
#assuming image is of square dimensions

def con_reflectance_pattern():
  img=np.zeros((1000,1000),int)
  for i in range(pr_adv.shape[0]):
    for j in range(pr_adv.shape[1]):
      img[i][j]=255
  #starting point for iterations
  if img.shape[0]%2==0 :
        a=img.shape[0]//2-1
        b=img.shape[0]//2-1
        x=img.shape[0]//2
        y=img.shape[0]//2
  else:
        a=img.shape[0]//2-1
        b=img.shape[0]//2-1
        x=img.shape[0]//2+1
        y=img.shape[0]//2+1
        
  loop_counter=0.51    
  while a!=-1 and b!=-1:
    for y_index in range(b,y+1):
      img[a][y_index]=img[a][y_index]-loop_counter
      img[x][y_index]=img[x][y_index]-loop_counter
      
    for x_index in range(a+1,x):
      img[x_index][b]=img[x_index][b]-loop_counter
      img[x_index][y]=img[x_index][y]-loop_counter
      
    a=a-1
    b=b-1
    x=x+1
    y=y+1        
    loop_counter=loop_counter+0.51
  plt.imshow(img,cmap='gray')  
con_reflectance_pattern()
```
